```python
import numpy as np

# PASSO 0: Definir as variáveis ​​necessárias para os cálculos de forma estática (ESTAS VARIÁVEIS DEVEM VIR DA BD OU DE CONSULTAS)
# LISTA DE ATIVIDADES:
actividades = [
  "Técnica de construção, pintura, colagem, recorte, guardanapo", "Trabalhar com recurso a materiais reciclados", "Execução de trabalhos para festas",
  "Trabalhos em feltro", "Pequenos arranjos", "Quadros", "Trabalhos para a decoração do Centro", "Confeção de sobremesas ou de sopas", "Exercício de orientação temporal, espacial e pessoal",
  "Exercícios de Conhecimento e realização de ações", "Exercícios de reconhecimento e manipulação espacial", "Exercícios de tarefas verbais sobre as partes do corpo",
  "Jogos de Reconhecimento da sua própria imagem", "Exercícios de memória recentes, imediata, episódica e semântica", "Jogos de destreza manual", "Exercícios de leitura e escrita",
  "Jogos das diferenças", "Exercícios de compreensão/conhecimento numérico e cálculo", "Jogos de memória", "Exercícios/jogos de reconhecimento dos sentidos",
  "Reagir aos diferentes sabores", "Diferenciar alguns cheiros", "Distinguir vozes de animais", "Caminhadas", "Exercícios de aquecimento e relaxamento", "Fisioterapia", "Karaoke", "Musicograma",
  "Utilização de instrumentos musicais", "Jogo de adivinhar músicas", "Petanca", "Bowling", "Lencinho adaptado", "Mímica", "Bingo", "Missako", "Dominó", "Jogo do Galo", "Manicure", "Massagens",
  "Maquilhagem", "Leitura de textos dramáticos", "Interpretação de personagens"
];

# CRITERIOS LIST
criterios = [
    'Estimulação cognitiva', 'Estimulação física', 'Facilidade de execução', 'Interação social', 
    'Controle de sintomas', 'Duração da atividade (hr)', 'Recursos necessários', 'Custo da atividade (euros)', 
    'Necessidade de deslocamento (0/1)', 'Nível de autonomia', 'Atractividade e prazer pessoal', 
    'Segurança', 'Pessoal cuidador (número)'
]

# PESOS PARA CADA DOENÇA
pesos_criterios_alzeimer = np.array([0.22, 0.05, 0.06, 0.15, 0.15, 0.03, 0.02, 0.03, 0.02, 0.02, 0.02, 0.21, 0.02]) #suma = 1
pesos_criterios_parkinson = np.array([0.06, 0.25, 0.08, 0.05, 0.15, 0.03, 0.02, 0.04, 0.02, 0.02, 0.02, 0.24, 0.02]) #suma = 1
pesos_criterios_ambos = np.array([0.16, 0.16, 0.08, 0.10, 0.12, 0.03, 0.02, 0.02, 0.02, 0.01, 0.05, 0.22, 0.01]) #suma = 1
pesos_criterios_ninguno = np.array([0.10, 0.08, 0.10, 0.10, 0.08, 0.05, 0.03, 0.04, 0.03, 0.06, 0.11, 0.19, 0.03]) #suma = 1

enfermedad_detectada = "Alzheimer"  # Dependendo da doença, selecionamos os pesos correspondentes

if enfermedad_detectada == "Alzheimer":
    weights = pesos_criterios_alzeimer
elif enfermedad_detectada == "Parkinson":
    weights = pesos_criterios_parkinson
elif enfermedad_detectada == "Ambos":
    weights = pesos_criterios_ambos
else:
    weights = pesos_criterios_ninguno

# ÍNDICES DAS COLUNAS A MAXIMIZAR E MINIMIZAR
criterios_maximizar = np.array([True, True, True, True, True, False, False, False, False, True, True, True, False])  # Ejemplo: el 1er y 3er criterio son de maximización, el 2do es de minimización

# DICCIONARIO PARA SUSTITUIR VALORES DE LA MATRIZ DE DECISION
reemplazos = {
    "Muy Alta": 5,
    "Alta": 4,
    "Moderada": 3,
    "Baja": 2,
    "Muy Baja": 1,
}

# PASSO 1: Definir a matriz de decisão
# Cada linha representa uma atividade e cada coluna um critério
# As atividades são: Técnica de construção, pintura, colagem, recorte, etc.
# Os critérios são: Estimulação cognitiva, Estimulação física, Facilidade de execução, etc.
decisiones =  np.array([
    ["Muy Alta", "Moderada", "Moderada", "Alta", "Moderada", 2, "Moderada", 12, 0, "Moderada", "Alta", "Alta", 1],
    ["Alta", "Moderada", "Moderada", "Alta", "Moderada", 2, "Baja", 8, 0, "Alta", "Alta", "Alta", 1],
    ["Alta", "Baja", "Moderada", "Muy Alta", "Baja", 3, "Baja", 10, 0, "Baja", "Alta", "Alta", 2],
    ["Moderada", "Moderada", "Moderada", "Muy Baja", "Moderada", 1.5, "Moderada", 15, 0, "Moderada", "Moderada", "Alta", 1],
    ["Moderada", "Moderada", "Alta", "Moderada", "Moderada", 1, "Baja", 5, 0, "Alta", "Alta", "Alta", 1],
    ["Muy Alta", "Moderada", "Moderada", "Baja", "Alta", 2.5, "Moderada", 20, 0, "Baja", "Alta", "Alta", 1],
    ["Alta", "Baja", "Moderada", "Alta", "Moderada", 3, "Moderada", 20, 0, "Moderada", "Alta", "Moderada", 2],
    ["Alta", "Moderada", "Moderada", "Muy Alta", "Moderada", 1.5, "Moderada", 15, 0, "Muy Baja", "Alta", "Moderada", 2],
    ["Alta", "Baja", "Alta", "Moderada", "Moderada", 0.5, "Baja", 5, 0, "Alta", "Moderada", "Alta", 1],
    ["Muy Alta", "Moderada", "Moderada", "Moderada", "Moderada", 0.5, "Muy Baja", 5, 0, "Moderada", "Moderada", "Alta", 1],
    ["Alta", "Moderada", "Moderada", "Moderada", "Moderada", 0.5, "Baja", 5, 0, "Moderada", "Alta", "Alta", 1],
    ["Alta", "Baja", "Moderada", "Moderada", "Moderada", 0.5, "Baja", 5, 0, "Alta", "Moderada", "Alta", 1],
    ["Alta", "Baja", "Moderada", "Moderada", "Moderada", 0.5, "Baja", 0, 0, "Moderada", "Alta", "Alta", 1],
    ["Muy Alta", "Baja", "Alta", "Moderada", "Moderada", 0.5, "Baja", 5, 0, "Moderada", "Alta", "Alta", 1],
    ["Moderada", "Alta", "Moderada", "Moderada", "Moderada", 0.5, "Baja", 5, 0, "Moderada", "Alta", "Alta", 1],
    ["Alta", "Baja", "Moderada", "Baja", "Moderada", 1, "Baja", 7, 0, "Moderada", "Moderada", "Alta", 1],
    ["Alta", "Baja", "Moderada", "Moderada", "Moderada", 0.5, "Baja", 5, 0, "Alta", "Alta", "Alta", 1],
    ["Alta", "Baja", "Moderada", "Baja", "Moderada", 1, "Baja", 5, 0, "Moderada", "Moderada", "Alta", 1],
    ["Muy Alta", "Baja", "Moderada", "Moderada", "Alta", 0.5, "Baja", 5, 0, "Alta", "Alta", "Alta", 1],
    ["Alta", "Baja", "Alta", "Moderada", "Moderada", 0.5, "Muy Baja", 5, 0, "Alta", "Alta", "Alta", 1],
    ["Moderada", "Baja", "Alta", "Moderada", "Moderada", 0.5, "Moderada", 10, 0, "Moderada", "Alta", "Alta", 1],
    ["Moderada", "Baja", "Alta", "Moderada", "Moderada", 0.5, "Baja", 8, 0, "Alta", "Alta", "Alta", 1],
    ["Moderada", "Baja", "Alta", "Moderada", "Moderada", 0.5, "Muy Baja", 5, 0, "Alta", "Alta", "Alta", 1],
    ["Moderada", "Alta", "Moderada", "Alta", "Alta", 1, "Baja", 5, 1, "Moderada", "Alta", "Moderada", 2],
    ["Moderada", "Moderada", "Alta", "Moderada", "Alta", 0.5, "Baja", 0, 0, "Alta", "Alta", "Alta", 1],
    ["Moderada", "Alta", "Moderada", "Baja", "Alta", 1, "Moderada", 20, 1, "Muy Baja", "Moderada", "Alta", 2],
    ["Alta", "Baja", "Moderada", "Alta", "Moderada", 1, "Baja", 5, 0, "Moderada", "Alta", "Muy Alta", 1],
    ["Alta", "Baja", "Moderada", "Moderada", "Moderada", 0.5, "Moderada", 10, 0, "Moderada", "Alta", "Muy Alta", 1],
    ["Alta", "Moderada", "Moderada", "Alta", "Moderada", 1, "Moderada", 10, 0, "Moderada", "Alta", "Alta", 1],
    ["Alta", "Baja", "Alta", "Moderada", "Moderada", 0.5, "Baja", 0, 0, "Alta", "Alta", "Alta", 1],
    ["Moderada", "Alta", "Moderada", "Alta", "Alta", 2, "Moderada", 5, 0, "Moderada", "Alta", "Moderada", 1],
    ["Moderada", "Muy Alta", "Moderada", "Alta", "Alta", 2, "Moderada", 5, 1, "Baja", "Alta", "Moderada", 2],
    ["Moderada", "Alta", "Alta", "Alta", "Moderada", 0.5, "Baja", 5, 0, "Moderada", "Alta", "Alta", 1],
    ["Alta", "Moderada", "Alta", "Alta", "Moderada", 1, "Baja", 0, 0, "Alta", "Alta", "Alta", 1],
    ["Alta", "Baja", "Alta", "Alta", "Moderada", 2, "Moderada", 8, 0, "Alta", "Alta", "Muy Alta", 3],
    ["Moderada", "Baja", "Alta", "Moderada", "Moderada", 0.5, "Baja", 5, 0, "Alta", "Moderada", "Alta", 1],
    ["Alta", "Baja", "Muy Alta", "Moderada", "Moderada", 1, "Baja", 0, 0, "Alta", "Alta", "Alta", 1],
    ["Alta", "Baja", "Muy Alta", "Moderada", "Moderada", 0.5, "Muy Baja", 0, 0, "Alta", "Alta", "Muy Alta", 0],
    ["Baja", "Baja", "Alta", "Moderada", "Moderada", 0.5, "Moderada", 10, 0, "Alta", "Alta", "Alta", 1],
    ["Baja", "Baja", "Alta", "Moderada", "Alta", 0.5, "Moderada", 5, 0, "Alta", "Alta", "Alta", 1],
    ["Baja", "Baja", "Moderada", "Moderada", "Moderada", 1, "Moderada", 20, 0, "Alta", "Alta", "Alta", 1],
    ["Alta", "Baja", "Moderada", "Moderada", "Moderada", 1, "Baja", 0, 0, "Alta", "Alta", "Alta", 1],
    ["Alta", "Moderada", "Moderada", "Alta", "Moderada", 1, "Baja", 0, 0, "Moderada", "Alta", "Alta", 1]
])

# Substituir sequencialmente as palavras da matriz pelos seus valores correspondentes
for i in range(decisiones.shape[0]):
    for j in range(decisiones.shape[1]):
        if isinstance(decisiones[i, j], str):  # Verificamos si es una cadena
            decisiones[i, j] = reemplazos.get(decisiones[i, j], decisiones[i, j])

# Convertemos a matriz em números
decisiones = decisiones.astype(float)

# PASSO 0: Normalizar a matriz de decisão
norma = np.sqrt(np.sum(decisiones**2, axis=0)) # Calcular a norma (raiz quadrada da soma dos quadrados por coluna)
matriz_normalizada = decisiones / norma

# Paso 1: Calcular las matrices de diferencia por criterio
def calcular_matrices_diferencia(D):
    m, n = D.shape
    diferencias_por_criterio = []

    # Calcular las diferencias por cada criterio
    for j in range(n):  # Para cada criterio (columna)
        diferencia_criterio = np.zeros((m, m))  # Inicializar la matriz de diferencia para este criterio
        for i in range(m):
            for k in range(m):
                if i != k:
                    diferencia_criterio[i, k] = D[i, j] - D[k, j]  # Diferencia por el criterio j
                else:
                    diferencia_criterio[i, k] = 0  # Las diagonales deben ser 0
        diferencias_por_criterio.append(diferencia_criterio)  # Añadir la matriz de diferencia del criterio j

    return diferencias_por_criterio

# Paso 2: Calcular la matriz de preferencia utilizando Tipo 1 Usual
def calcular_matriz_preferencia(diferencias, criterios_maximizar):
    m = diferencias[0].shape[0]  # Número de alternativas
    preferencias_por_criterio = []

    for idx, diferencia in enumerate(diferencias):
        preferencia_criterio = np.zeros((m, m))
        for i in range(m):
            for j in range(m):
                if criterios_maximizar[idx]:  # Maximización
                    if diferencia[i, j] > 0:
                        preferencia_criterio[i, j] = 1
                    else:
                        preferencia_criterio[i, j] = 0
                else:  # Minimización
                    if diferencia[i, j] < 0:
                        preferencia_criterio[i, j] = 1
                    else:
                        preferencia_criterio[i, j] = 0
        preferencias_por_criterio.append(preferencia_criterio)

    return preferencias_por_criterio

# Paso 3: Calcular el índice de preferencia ponderado
def calcular_indice_preferencia(preferencias, weights):
    m = preferencias[0].shape[0]  # Número de alternativas
    indice_preferencia = np.zeros((m, m))  # Matriz de índice de preferencia

    # Sumar las matrices de preferencia ponderadas
    for idx, preferencia in enumerate(preferencias):
        indice_preferencia += weights[idx] * preferencia  # Multiplicar cada matriz de preferencia por el peso correspondiente

    return indice_preferencia
    
# Paso 4: Calcular los flujos de superación
def calcular_flujos_superacion(indice_preferencia):
    m = indice_preferencia.shape[0]
    flujo_superacion_positivo = np.zeros(m)
    flujo_superacion_negativo = np.zeros(m)

    # Normalizamos por m-1 (en lugar de simplemente sumar)
    for i in range(m):
        flujo_superacion_positivo[i] = np.sum(indice_preferencia[i, :]) / (m - 1)
        flujo_superacion_negativo[i] = np.sum(indice_preferencia[:, i]) / (m - 1)

    # Calcular flujo neto de superación (Φ)
    flujo_neto_superacion = flujo_superacion_positivo - flujo_superacion_negativo

    return flujo_superacion_positivo, flujo_superacion_negativo, flujo_neto_superacion

# Calcular matrices de diferencias
diferencias = calcular_matrices_diferencia(matriz_normalizada)
# Calcular matrices de preferencias
preferencias = calcular_matriz_preferencia(diferencias, criterios_maximizar)
# Calcular matrices de preferencias ponderadas
indice_preferencia = calcular_indice_preferencia(preferencias, weights)
# Calcular flujos positivos, negativos, netos
flujo_superacion_positivo, flujo_superacion_negativo, flujo_neto_superacion = calcular_flujos_superacion(indice_preferencia)

# Asociar los flujos netos a las alternativas (índices)
flujos_con_alternativas = list(zip(actividades, flujo_neto_superacion))
flujos_con_alternativas = [(i + 1, actividad, flujo) for i, (actividad, flujo) in enumerate(zip(actividades, flujo_neto_superacion))]
# Ordenar los flujos netos de mayor a menor
flujos_ordenados = sorted(flujos_con_alternativas, key=lambda x: x[2], reverse=True)

# Mostrar los flujos netos ordenados
print("\nRanking de Alternativas (PROMETHEE II):")
print("Flujos Netos de Superación (Φ) Ordenados de Mayor a Menor:")
print("\nAlternativa | Actividad | Flujo Neto")
for alt, actividad, flujo in flujos_ordenados:
    print(f"alt {alt:2d} | {actividad}: {flujo:.4f}")
```
